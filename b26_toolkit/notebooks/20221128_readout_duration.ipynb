## normalizing the signal
deviding by time

```python
plt.plot(data["tau"],data["counts"][:,0]/data["tau"],label="reference")
plt.plot(data["tau"],data["counts"][:,1]/data["tau"],label="signal")
plt.legend()
plt.xlabel(r'$\tau\,[ns]$')
plt.ylabel("accumalated counts")
```

## calculating the contrast

```python
ref = data["counts"][:,0]
sig = data["counts"][:,1]
tau = data["tau"]

contrast = (ref-sig)/(ref+sig)
plt.plot(tau,contrast)
```
